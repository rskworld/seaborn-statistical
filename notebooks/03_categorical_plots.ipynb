This notebook demonstrates categorical and relational plots:
- Bar plots
- Count plots
- Box plots
- Violin plots
- Swarm plots
- Strip plots


```python
# Author: Molla Samser
# Website: https://rskworld.in/
# Email: help@rskworld.in
# Phone: +91 93305 39277
# Address: Nutanhat, Mongolkote, Purba Burdwan, West Bengal, India, 713147

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_style("whitegrid")
sns.set_palette("Set2")
plt.rcParams['figure.figsize'] = (12, 6)

# Load data
df = pd.read_csv('../data/sample_data.csv')
print("Dataset shape:", df.shape)
df.head()
```

## 1. Bar Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Bar plot - average income by category
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='category', y='income', estimator='mean', ci='sd')
plt.title('Average Income by Category', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Income', fontsize=12)
plt.tight_layout()
plt.savefig('../images/barplot_category.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Bar plot with hue
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='category', y='income', hue='gender', estimator='mean')
plt.title('Average Income by Category and Gender', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Income', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.tight_layout()
plt.savefig('../images/barplot_hue.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 2. Count Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Count plot
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(data=df, x='category', ax=axes[0])
axes[0].set_title('Count by Category', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Category', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)

sns.countplot(data=df, x='gender', hue='category', ax=axes[1])
axes[1].set_title('Count by Gender and Category', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Gender', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].legend(title='Category', fontsize=9)

plt.tight_layout()
plt.savefig('../images/countplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 3. Box Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='category', y='income', hue='gender')
plt.title('Income Distribution by Category and Gender (Box Plot)', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.tight_layout()
plt.savefig('../images/boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 4. Violin Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Violin plot - shows distribution shape
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='category', y='score', hue='gender', split=True)
plt.title('Score Distribution by Category and Gender (Violin Plot)', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.tight_layout()
plt.savefig('../images/violinplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 5. Swarm Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Swarm plot - shows individual data points
plt.figure(figsize=(12, 6))
sns.swarmplot(data=df, x='category', y='sales', hue='gender', size=4)
plt.title('Sales Distribution by Category (Swarm Plot)', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.tight_layout()
plt.savefig('../images/swarmplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 6. Combined Box and Swarm Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Combine box plot with swarm plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='category', y='income', color='lightgray')
sns.swarmplot(data=df, x='category', y='income', hue='gender', size=4)
plt.title('Income Distribution: Box Plot + Swarm Plot', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.tight_layout()
plt.savefig('../images/box_swarm_combined.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 7. Strip Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Strip plot
plt.figure(figsize=(12, 6))
sns.stripplot(data=df, x='region', y='income', hue='category', jitter=True, size=5)
plt.title('Income Distribution by Region (Strip Plot)', fontsize=16, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.legend(title='Category', fontsize=10)
plt.tight_layout()
plt.savefig('../images/stripplot.png', dpi=300, bbox_inches='tight')
plt.show()
```
