This notebook demonstrates various distribution plots using Seaborn:
- Histograms
- KDE (Kernel Density Estimation) plots
- Rug plots
- Distribution comparisons


```python
# Author: Molla Samser
# Website: https://rskworld.in/
# Email: help@rskworld.in
# Phone: +91 93305 39277
# Address: Nutanhat, Mongolkote, Purba Burdwan, West Bengal, India, 713147

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_style("whitegrid")
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

# Load data
df = pd.read_csv('../data/sample_data.csv')
print("Dataset shape:", df.shape)
print("\nFirst few rows:")
df.head()
```

## 1. Histogram Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Basic histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='income', bins=20, kde=True)
plt.title('Income Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Income', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.savefig('../images/income_histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Histogram with multiple variables
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(data=df, x='age', bins=15, kde=True, ax=axes[0])
axes[0].set_title('Age Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Age', fontsize=12)

sns.histplot(data=df, x='score', bins=20, kde=True, ax=axes[1])
axes[1].set_title('Score Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Score', fontsize=12)

plt.tight_layout()
plt.savefig('../images/multiple_histograms.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 2. KDE (Kernel Density Estimation) Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='income', fill=True, alpha=0.7)
plt.title('Income KDE Plot', fontsize=16, fontweight='bold')
plt.xlabel('Income', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.tight_layout()
plt.savefig('../images/income_kde.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Multiple KDE plots by category
plt.figure(figsize=(12, 6))
for category in df['category'].unique():
    sns.kdeplot(data=df[df['category'] == category], x='income', 
                label=category, fill=True, alpha=0.5)

plt.title('Income Distribution by Category', fontsize=16, fontweight='bold')
plt.xlabel('Income', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Category', fontsize=10)
plt.tight_layout()
plt.savefig('../images/kde_by_category.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 3. Rug Plot


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# KDE plot with rug plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='score', fill=True, alpha=0.6)
sns.rugplot(data=df, x='score', height=0.05)
plt.title('Score Distribution with Rug Plot', fontsize=16, fontweight='bold')
plt.xlabel('Score', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.tight_layout()
plt.savefig('../images/rug_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 4. Distribution Comparison


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# Compare distributions by gender
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(data=df, x='income', hue='gender', kde=True, ax=axes[0])
axes[0].set_title('Income Distribution by Gender', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Income', fontsize=12)

sns.kdeplot(data=df, x='income', hue='gender', fill=True, alpha=0.6, ax=axes[1])
axes[1].set_title('Income KDE by Gender', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Income', fontsize=12)

plt.tight_layout()
plt.savefig('../images/distribution_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 5. Bivariate Distribution


```python
# Author: Molla Samser
# Website: https://rskworld.in/

# 2D KDE plot
plt.figure(figsize=(10, 8))
sns.kdeplot(data=df, x='age', y='income', fill=True, cmap='viridis', alpha=0.7)
plt.title('Age vs Income Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.tight_layout()
plt.savefig('../images/bivariate_kde.png', dpi=300, bbox_inches='tight')
plt.show()
```
